## Tensorflow

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image_dataset_from_directory

train_dir = "/home/dell/Documentos/Untitled Folder/cats_dogs/PetImages/train/"
validation_dir = "/home/dell/Documentos/Untitled Folder/cats_dogs/PetImages/val/"
test_dir = "/home/dell/Documentos/Untitled Folder/cats_dogs/PetImages/test/"

training_set = image_dataset_from_directory(train_dir,
                                             shuffle=True,
                                             labels="inferred",
                                             label_mode="int",
                                             batch_size=32,
                                             image_size=(150, 150))

val_dataset = image_dataset_from_directory(validation_dir,
                                                  shuffle=True,
                                                  labels="inferred",
                                                  label_mode="int",
                                                  batch_size=32,
                                                  image_size=(150, 150))

test_dataset = image_dataset_from_directory(test_dir,
                                                  shuffle=True,
                                                  labels="inferred",
                                                  label_mode="int",
                                                  batch_size=32,
                                                  image_size=(150, 150))

os.chdir('/home/dell/Documentos/Untitled Folder/')
```

## Data pre-processing



```python
data_augmentation = keras.Sequential(
    [       keras.layers.experimental.preprocessing.RandomFlip("horizontal"),
   keras.layers.experimental.preprocessing.RandomRotation(0.1),
    ]
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

for images, labels in training_set.take(1):
    plt.figure(figsize=(12, 12))
    first_image = images[0]
    for i in range(12):
        ax = plt.subplot(3, 4, i + 1)
        augmented_image = data_augmentation(
            tf.expand_dims(first_image, 0)
        )
        plt.imshow(augmented_image[0].numpy().astype("int32"))
        plt.axis("off")
```

## Modelo base

```python
base_model = keras.applications.Xception(
    weights='imagenet',  
    input_shape=(150, 150, 3),
    include_top=False)  

base_model.trainable = False
```

```python
inputs = keras.Input(shape=(150, 150, 3))
x = data_augmentation(inputs) 
x = tf.keras.applications.xception.preprocess_input(x)
```

```python
x = base_model(x, training=False)
x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dropout(0.2)(x)  
outputs = keras.layers.Dense(1)(x)

model = keras.Model(inputs, outputs)
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),metrics=keras.metrics.BinaryAccuracy())

model.fit(training_set, epochs=20, validation_data=val_dataset)
```

## Fine Tuning

```python
base_model.trainable = True

model.compile(optimizer=keras.optimizers.Adam(1e-5),  
              loss=keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=keras.metrics.BinaryAccuracy())


# prevent overfit

from tensorflow.keras.callbacks import EarlyStopping, TensorBoard

#rm -rf logs
%load_ext tensorboard
log_folder = 'logs'
callbacks = [
            EarlyStopping(patience = 5),
            TensorBoard(log_dir=log_folder)
            ]
# model fit

model.fit(training_set, epochs=15,validation_data=val_dataset,callbacks=callbacks)
```
